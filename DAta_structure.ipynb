```python
                                              2. stack  
                                        How does this data structure technique store the data?

Stacks store data in a sequential manner, following the Last In, First Out (LIFO) principle. Elements are added and removed from the top of the stack.
Each time an element is added (pushed) onto the stack, it becomes the new top element. Similarly, when an element is removed (popped) from the stack, 
the element that was added last is the one that gets removed.

                                         Why is it stored in that way?
Storing data in a stack-like manner facilitates efficient last-in, and first-out operations, which are common in many scenarios. 
This storage approach is particularly useful in situations where you need to keep track of a sequence of elements and where the most recent elements are the ones of immediate interest.
For example, in function call management, the most recently called function needs to finish executing before the previous one can resume, making a stack an ideal data structure.

                                         What are the scenarios we use it?
Stacks are used in various scenarios, including:

Function call management and recursion in programming languages.
Expression evaluation and conversion (infix to postfix/prefix).
Backtracking algorithms like Depth-First Search (DFS).
Undo functionality in applications.
Syntax parsing and checking in compilers.
Memory management, particularly in managing function call frames and local variables.

History management in web browsers, where the back and forward buttons simulate a stack-like behavior.
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head is None
    
    def push(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
    
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        else:
            popped_item = self.head.data
            self.head = self.head.next
            return popped_item
    
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        else:
            return self.head.data
    
    def display(self):
        current = self.head
        if self.is_empty():
            print("Stack is empty")
        else:
            while current is not None:
                print(current.data, end=" -> ")
                current = current.next
            print("None")

# Example usage:
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.display()  

print("Popped:", stack.pop())  
stack.display()  

print("Peeked:", stack.peek()) 
stack.display() 
```

                                           3 queue
                                           How does this data structure technique store the data?

A queue stores data in a First-In-First-Out (FIFO) manner. This means that the first element added to the queue is the first one to be removed. In terms of implementation, a queue can be implemented using arrays/lists or linked lists. In the case of a linked list implementation, each element of the queue is stored in a node, and these nodes are linked together                                          .
Why is it stored in that way?

Storing data in a FIFO manner is useful in scenarios where you need to process data in the order it was received. For example, in a printer queue, documents are printed in the order they were sent to the printer. This ensures fairness and efficiency in processing ta                                         sks.
Scenarios and use cas1. uch as:
Task scheduling: A queue can be used to schedule tasks in the order they are r2. eceived.
Printers: A print queue manages the order in which documents are3.  printed.
Message queues: In distributed systems, message queues ensure that messages are processed in the order they 4. were sent.
Breadth-first search (BFS) traversal in graphs: A queue is used to traverse nodes level by level in BFS

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return None
        else:
            popped_item = self.front.data
            if self.front == self.rear:
                self.front = self.rear = None
            else:
                self.front = self.front.next
            return popped_item

    def display(self):
        current = self.front
        if self.is_empty():
            print("Queue is empty")
        else:
            while current is not None:
                print(current.data, end=" -> ")
                current = current.next
            print("None")

# Example usage:
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.display()  # Output: 1 -> 2 -> 3 -> None

print("Dequeued:", queue.dequeue())  # Output: Dequeued: 1
queue.display()  # Output: 2 -> 3 -> None
```

A deque, short for double-ended queue, is a data structure that allowthe s insertion and deletion of elements from both the front and the back. Storing data in a deque allows for efficient insertion and deletion operations at both ends of the data structure. This flexibility makes deques suitable for scenarios where elements need to be added or removed from both ends frequently. For example, deques are commonly used in scenarios such as implementing queues, stacks, and double-ended priority qu

Deques are commonly used in scenarios where elements need to be added or removed from both ends, such as:

Sliding window problems: Deques are useful in solving problems where you need to maintain a window of elements and perform operations on it efficiently, such as finding the maximum or minimum in a sliding window of fixed size.
Palindrome checking: Deques are used to efficiently check if a string is a palindrome by comparing characters from

```python
from collections import deque

class MyDeque:
    def __init__(self):
        self.deque = deque()

    def is_empty(self):
        return len(self.deque) == 0

    def add_front(self, data):
        self.deque.appendleft(data)

    def add_rear(self, data):
        self.deque.append(data)

    def remove_front(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque.popleft()

    def remove_rear(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque.pop()

    def peek_front(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque[0]

    def peek_rear(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque[-1]

    def size(self):
        return len(self.deque)

    def display(self):
        if self.is_empty():
            print("Deque is empty")
        else:
            print("Front ->", end=" ")
            for item in self.deque:
                print(item, end=" ")
            print("<- Rear")

# Example usage:
my_deque = MyDeque()
my_deque.add_front(1)
my_deque.add_rear(2)
my_deque.add_front(3)
my_deque.display()  # Output: Front -> 3 1 2 <- Rear

print("Removed from front:", my_deque.remove_front())  # Output: Removed from front: 3
print("Removed from rear:", my_deque.remove_rear())    # Output: Removed from rear: 2
my_deque.display()  # Output: Front -> 1 <- Rear

print("Peek front:", my_deque.peek_front())  # Output: Peek front: 1
print("Peek rear:", my_deque.peek_rear())    # Output: Peek rear: 1

print("Deque size:", my_deque.size())  # Output: Deque size: 1

from collections import deque

class MyDeque:
    def __init__(self):
        self.deque = deque()

    def is_empty(self):
        return len(self.deque) == 0

    def add_front(self, data):
        self.deque.appendleft(data)

    def add_rear(self, data):
        self.deque.append(data)

    def remove_front(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque.popleft()

    def remove_rear(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque.pop()

    def peek_front(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque[0]

    def peek_rear(self):
        if self.is_empty():
            print("Deque is empty")
            return None
        else:
            return self.deque[-1]

    def size(self):
        return len(self.deque)

    def display(self):
        if self.is_empty():
            print("Deque is empty")
        else:
            print("Front ->", end=" ")
            for item in self.deque:
                print(item, end=" ")
            print("<- Rear")

# Example usage:
my_deque = MyDeque()
my_deque.add_front(1)
my_deque.add_rear(2)
my_deque.add_front(3)
my_deque.display()  # Output: Front -> 3 1 2 <- Rear

print("Removed from front:", my_deque.remove_front())  # Output: Removed from front: 3
print("Removed from rear:", my_deque.remove_rear())    # Output: Removed from rear: 2
my_deque.display()  # Output: Front -> 1 <- Rear

print("Peek front:", my_deque.peek_front())  # Output: Peek front: 1
print("Peek rear:", my_deque.peek_rear())    # Output: Peek rear: 1

print("Deque size:", my_deque.size())  # Output: Deque size: 1



```
